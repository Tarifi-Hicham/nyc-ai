## Manhatten & Euclidian distances
---

```python
#create function to calculate Manhattan distance 
def manhattan(a, b):
    return sum(abs(val1-val2) for val1, val2 in zip(a,b))

#create function to calculate Euclidean distance
def euclidean(a, b):
    return sqrt(sum((val1-val2)**2 for val1, val2 in zip(a,b)))
```

```python
# Apply manhattan distance to the test data
df_test['manhattan'] = df_test.apply(lambda row: manhattan(row[4:5], row[6:7]), axis=1)
df_test.head()
```
